#### Measure of dispersion
1) Variance
2) Standard Deviation
   

```python
statistics.variance(ages1) ## for sample 
```

```python
print(np.var(ages1,axis=0)) ## For population
```

```python
### Manually By using Statistical Fromula 

def variance(data):
    n=len(ages1)
    mean=sum(data)/n
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/n
    return variance
```

```python
variance(ages1)
```

```python
def variance(data):
    n=len(ages1)
    mean=sum(data)/n
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-1)
    return variance
```

```python
variance(ages1)
```

```python
def variance(data,dof=0):
    n=len(ages1)
    mean=sum(data)/n
    deviation=[(x-mean)**2 for x in data]
    variance=sum(deviation)/(n-dof)
    return variance
```

```python
variance(ages1,dof=1)
```

#### Check whether Distribution is normal Distribution

```python
#### Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
    
```

```python
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```
